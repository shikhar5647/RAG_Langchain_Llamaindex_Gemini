```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings

Settings.embed_model = HuggingFaceEmbedding(
    model_name="BAAI/bge-small-en-v1.5"
)
```

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms.gemini import Gemini
llm=Gemini()
data = SimpleDirectoryReader(input_dir="./data/paul_graham/").load_data()
index = VectorStoreIndex.from_documents(data)
```

## Best Mode

```python
chat_engine = index.as_chat_engine(chat_mode="best", llm=llm, verbose=True)
```

```python
response = chat_engine.chat(
    "What are the first programs Paul Graham tried writing?"
)
```

```python
print(response)
```

## Condense plus Context

```python
from llama_index.core.memory import ChatMemoryBuffer

memory = ChatMemoryBuffer.from_defaults(token_limit=3900)

chat_engine = index.as_chat_engine(
    chat_mode="condense_plus_context",
    memory=memory,
    llm=llm,
    context_prompt=(
        "You are a chatbot, able to have normal interactions, as well as talk"
        " about an essay discussing Paul Grahams life."
        "Here are the relevant documents for the context:\n"
        "{context_str}"
        "\nInstruction: Use the previous chat history, or the context above, to interact and help the user."
    ),
    verbose=False,
)
```

```python
response = chat_engine.chat("What did Paul Graham do growing up")
```

```python
print(response)
```

```python
response_2 = chat_engine.chat("Can you tell me more?")
```

```python
print(response_2)
```

```python
chat_engine.reset()
```

```python
response = chat_engine.chat("Hello! What do you know?")
```

```python
print(response)
```

## Chat Engine with Personality

```python
from llama_index.core.chat_engine import SimpleChatEngine

chat_engine = SimpleChatEngine.from_defaults(llm=llm)
response = chat_engine.chat(
    "Say something profound about Jodhpur"
)
print(response)
```

```python
from llama_index.core.prompts.system import SHAKESPEARE_WRITING_ASSISTANT
chat_engine = SimpleChatEngine.from_defaults(llm=llm,
    system_prompt=SHAKESPEARE_WRITING_ASSISTANT
)
response = chat_engine.chat(
    "Say something profound and romantic about Jodhpur"
)
print(response)
```

```python
from llama_index.core.prompts.system import MARKETING_WRITING_ASSISTANT
chat_engine = SimpleChatEngine.from_defaults(llm=llm,
    system_prompt=MARKETING_WRITING_ASSISTANT
)
response = chat_engine.chat(
    "Say something profound about Jodhpur"
)
print(response)
```

```python
from llama_index.core.chat_engine import SimpleChatEngine
from llama_index.core.prompts.system import IRS_TAX_CHATBOT

chat_engine = SimpleChatEngine.from_defaults(llm=llm,system_prompt=IRS_TAX_CHATBOT)
response = chat_engine.chat(
    "Say something profound about Jodhpur"
)
print(response)
```
